# Graphs for Practical 2

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Graph generation


# Load CSV data
file_path = "results/ingest_benchmark_results_total.csv"
df = pd.read_csv(file_path)

# Compute averages
avg_ingestion = [df['Redis Ingestion'].mean(), df['ChromaDB Ingestion'].mean(), df['MongoDB Ingestion'].mean()]
avg_query = [df['Redis Query'].mean(), df['ChromaDB Query'].mean(), df['MongoDB Query'].mean()]
avg_storage = [df['Redis Storage (MB)'].mean(), df['ChromaDB Storage (MB)'].mean(), df['MongoDB Storage (MB)'].mean()]

# Without embedding
df_filtered = df[df['Embedding Model'] != 'sentence-transformers/all-MiniLM-L6-v2']
avg_storage_filtered = [df_filtered['Redis Storage (MB)'].mean(), df_filtered['ChromaDB Storage (MB)'].mean(), df_filtered['MongoDB Storage (MB)'].mean()]

# Labels
labels = ['Redis', 'ChromaDB', 'MongoDB']

# Plotting
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Ingestion Time
axes[0].bar(labels, avg_ingestion, color=['grey', 'olive', 'brown'])
axes[0].set_title('Average Ingestion Time (s)')
axes[0].set_ylabel('Time (Seconds)')

# Query Time
axes[1].bar(labels, avg_query, color=['grey', 'olive', 'brown'])
axes[1].set_title('Average Query Time (s)')
axes[1].set_ylabel('Time (Seconds)')

# Storage Size
axes[2].bar(labels, avg_storage, color=['grey', 'olive', 'brown'])
axes[2].set_title('Average Storage Size (MB)')
axes[2].set_ylabel('Storage (MB)')

# Storage Sizes 2
axes[3].bar(labels, avg_storage_filtered, color=['grey', 'olive', 'brown'])
axes[3].set_title('Storage Size (MB) without All-MiniLM-L6-v2')
axes[3].set_ylabel('Storage (MB)')

plt.tight_layout()
plt.show()
```
